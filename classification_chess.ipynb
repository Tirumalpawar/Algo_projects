```python
import pandas as pd
```

```python
df = pd.read_csv('chess_king_rook_dataset.csv')
df.head()
```

```python
df.info()
```

```python
df.keys()
```

```python
df['result'].value_counts()
```

```python
df["result"] = df["result"].map({'fourteen':0,'thirteen':1,'twelve':2,'eleven':3,'draw':4,'fifteen':5,'ten':6,'nine':7,
                                   'eight':8,'seven':9,'six':10,'five':11,'sixteen':12,'two':13,'four':14,'three':13,
                                  'one':14,'zero':15})
```

```python
df1 = pd.get_dummies(df,columns=['white_king_file','white_rook_file','black_king_file'],drop_first=True)
```

```python
df2 = pd.concat((df,df1),axis=1)
```

```python
df2.head()
```

```python
df2.drop(['white_king_file','white_rook_file','black_king_file','result'],axis=1,inplace=True)
```

```python
df2.head()
```

```python
df2.keys()
```

```python
X = df2.values
y = df['result'].values
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import VotingClassifier
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
lr = LogisticRegression()
knn = KNeighborsClassifier(n_neighbors=15)
dtc = DecisionTreeClassifier()
soft_voting = VotingClassifier
```

```python
soft_voting = VotingClassifier(estimators=[["log_reg",lr],["DTC",dtc],["KNC",knn]],voting="soft")
```

```python
soft_voting.fit(X_train,y_train)
```

```python
y_pred = soft_voting.predict(X_test)
```

```python
soft_voting.score(X_test,y_test)
```
