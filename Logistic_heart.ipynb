```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix,r2_score,classification_report
```

```python
df = pd.read_csv('Heart.csv')
df.head()
```

```python
df.info()
```

```python
df1 = df.fillna(0)
```

```python
df1.info()
```

```python
df1.keys()
```

```python
df1.head()
```

```python
chestpain = pd.get_dummies(df1['ChestPain'],drop_first=True)
thal = pd.get_dummies(df1['Thal'],drop_first=True)
ahd = pd.get_dummies(df1['AHD'],drop_first=True)
df2 = pd.concat([df1,chestpain,thal,ahd],axis=1)
```

```python
df2.head()
```

```python
df2.drop(['Thal','ChestPain','AHD'],axis=1,inplace=True)
```

```python
df2.head()
```

```python
X = df2.drop('Yes',axis=1)
```

```python
y = df2.iloc[:,-1]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
lrr = LogisticRegression()
```

```python
lrr.fit(X_train,y_train)
```

```python
y_pred = lrr.predict(X_test)
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```
