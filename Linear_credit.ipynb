```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, confusion_matrix, classification_report
```

```python
df = pd.read_csv('Credit')
df.head(5)
```

```python
df.info()
```

```python
df.keys()
```

```python
df.isnull()
```

```python
df1 = df.dropna()
```

```python
df1.info()
```

```python
gender = pd.get_dummies(df1['Gender'],drop_first=True)
student = pd.get_dummies(df1['Student'],drop_first=True)
married = pd.get_dummies(df1['Married'],drop_first=True)
ethnicity = pd.get_dummies(df1['Ethnicity'],drop_first=True)
```

```python
df2 = pd.concat([df1,gender,student,married,ethnicity],axis=1)
```

```python
df2.head()
```

```python
X = df2.iloc[:,[0,1,3,4,5,6,11,12,13,14,15,16]].values
y = df2.iloc[:,2].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=33)
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
print(r2_score(y_test,y_pred))
```
